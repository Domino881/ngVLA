```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open('MID_dist_metres.txt', 'r')

lines = f.readlines()

f.close()
```

```python
# Looking at the text file
for line in lines:
    print(line)
```

```python
dist_from_centre_total = []
dist_from_centre_MeerKAT = []
dist_from_centre_SKA1 = []

for line in lines:
    
    info = line.split()
    dist = (float(info[2]) ** 2 + float(info[3]) ** 2) ** 0.5 / 1000
    dist_from_centre_total.append(dist)
    
    if info[1].startswith('M'):
        dist_from_centre_MeerKAT.append(dist)

    elif info[1].startswith('SKA'):
        dist_from_centre_SKA1.append(dist)        
```

```python
dist_from_centre_total.sort()
dist_from_centre_MeerKAT.sort()
dist_from_centre_SKA1.sort()
```

```python
num_dishes_total = np.arange(len(dist_from_centre_total))
num_dishes_MeerKAT = np.arange(len(dist_from_centre_MeerKAT))
num_dishes_SKA1 = np.arange(len(dist_from_centre_SKA1))
```

```python
plt.plot(dist_from_centre_total, num_dishes_total, color='black', ls='-', lw=3, label='SKA1-Mid total')
plt.plot(dist_from_centre_MeerKAT, num_dishes_MeerKAT, color='red', ls='-', lw=3, label='MeerKAT dishes')
plt.plot(dist_from_centre_SKA1, num_dishes_SKA1, color='royalblue', ls='-', lw=3, label='SKA1 dishes')

plt.xscale('log')

plt.xlabel('Distance from array centre (km)')
plt.ylabel('Number of dishes')
plt.title('Size of SKA1-Mid array with respect to number of dishes')

plt.gcf().set_size_inches(15, 10)
plt.rc('font', size=15)
plt.grid()
plt.legend()
plt.show()
```

```python
f = open('ngvla-xyz.txt', 'r')

lines = f.readlines()

f.close()
```

```python
# Looking at the text file
for line in lines:
    print(line)
```

```python
ref_x, ref_y, ref_z = lines[129].split()[:3]
```

```python
dist_from_centre_total_old = []

for i, line in enumerate(lines):
    
    info = line.split()
    dist = ((float(info[0]) - float(ref_x)) ** 2 + \
    (float(info[1]) - float(ref_y)) ** 2 + \
    (float(info[2]) - float(ref_z)) ** 2) ** 0.5 / 1000
    dist_from_centre_total_old.append(dist)
```

```python
dist_from_centre_total_old.sort()
```

```python
num_dishes_total_old = np.arange(len(dist_from_centre_total_old))
```

```python
f = open('Rev_D_Config_Data/rev_d_xyz.txt', 'r')

lines = f.readlines()

f.close()
```

```python
# Looking at the text file
for line in lines:
    print(line)
```

```python
ref_x, ref_y, ref_z = lines[12].split()[:3]
```

```python
dist_from_centre_total_new = []

for i, line in enumerate(lines):
    
    info = line.split()
    dist = ((float(info[0]) - float(ref_x)) ** 2 + \
    (float(info[1]) - float(ref_y)) ** 2 + \
    (float(info[2]) - float(ref_z)) ** 2) ** 0.5 / 1000
    dist_from_centre_total_new.append(dist)
```

```python
dist_from_centre_total_new.sort()
```

```python
num_dishes_total_new = np.arange(len(dist_from_centre_total_new))
```

```python
plt.rc('font', size=15)

plt.plot(dist_from_centre_total_new, num_dishes_total_new, color='red', ls='-', lw=3, label='New config')
plt.plot(dist_from_centre_total_old, num_dishes_total_old, color='blue', ls='-', lw=3, label='Old config')

plt.xscale('log')

plt.xlabel('Distance from array centre (km)')
plt.ylabel('Number of dishes')
plt.title('Size of ngVLA array with respect to number of dishes')

plt.gcf().set_size_inches(15, 10)
plt.grid()
plt.legend()
plt.show()
```
